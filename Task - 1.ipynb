```python
import matplotlib.pyplot as plt

# Sample data
genders = ['Male', 'Female', 'Non-Binary', 'Other']
count = [80, 140, 40, 25]  # Replace with your actual data

# Create a bar chart
plt.bar(genders, count, color=['Red', 'yellow', 'green', 'orange'])

# Add labels and title
plt.xlabel('Genders')
plt.ylabel('Count')
plt.title('Distribution of Genders in a Population')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
ages = np.random.normal(loc=30, scale=5, size=900)  # Replace with your actual data

# Create a histogram
plt.hist(ages, bins=30, color='orange', edgecolor='black')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages in a Population')

# Show the plot
plt.show()
```
